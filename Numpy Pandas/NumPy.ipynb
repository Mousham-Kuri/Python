```python
import time
SIZE= 1000000

A1 = range(SIZE)
A2 = range(SIZE)

start= time.time()

result=[x+y for x,y in zip(A1,A2)]

stop=time.time()

print('python took ',(stop-start)*1000, ' ms to complete')


L1 = np.arange(SIZE)
L2 = np.arange(SIZE)

start = time.time()
result= L1+L2
stop=time.time()

print('numpy took ',(stop-start)*1000, ' ms to complete')

time.time()
```
