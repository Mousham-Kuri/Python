## Trending Video Analysis

**Filter those videos that were trending on the same day, find out the count of such videos**

```python
same_day = df[df['publish_date'] == df['trending_date']]
same_day.shape
```

**Find out the channel wise count of videos using same_day data and report which all channel has the maximum same day trending videos?**

```python
same_day.channel_title.value_counts()[:20]
```

**Find out the sum of views using same day data for various channels**

```python
same_day.groupby("channel_title").agg({'views':'sum', 'video_id':'count'}).sort_values(by = 'views', ascending = False).head(20)
```

Observations:

1. The top-4 channels which has the maximum same day trending videos belongs to Education genre.
2. Frequent Live Streama and Frequent uploads.


**Doubt**

Are they getting good views also?

The reason these channels are having so many same day trending videos is because, they are able to get significant views in less time by doing live stream which helps them to get their video in trending list but that's not a gurentee that the video will also get good viwership
